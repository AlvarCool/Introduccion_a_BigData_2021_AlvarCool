```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_html("https://finance.yahoo.com/quote/FB/history?p=FB")[0]
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.head(20).sort_values(by='Date')
```

```python
df
```

```python
df = df.drop([0, 0])
```

```python
df = df.astype({"Open":'float',
                "High":'float',
                "Low":'float',
                "Close*":'float',
                "Adj Close**":'float',
                "Volume":'float'})
```

```python
df
```

```python
df.corr()
```

```python
df['Gain'] = df['Close*'] - df['Open']
```

```python
df
```

```python
relation = df[['Open','High','Low']]
```

```python
relation
```

```python
relation.head(5)
```

```python
sns.pairplot(relation, kind = 'reg')
```
